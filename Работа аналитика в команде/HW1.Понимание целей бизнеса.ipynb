### Выбрать любой-интернет магазин и на основе теории AARRR продумать все важные метрики бизнеса, за которыми вы как собственник стали бы следить в первую очередь

AARRR
Acquisition - Привлечение
Activation - Активация
Retention - Удержание
Revenue - Доход
Referral - Рекомендация

Для примера берем OZON.ru

1.Привлечение
- Кол-во посещений сайта с уникальных ID/устройств
- Кол-во регистраций на сайте
- Кол-во незавершенных регистраций
- Динамика этих трех показателей по дням/неделям/месяцам
- Регистрации/посещение по источникам привлечения, источникам перехода на сайт

2.Активация
- Кол-во клиентов, совершивших первую сделку
- Дельта по времени между регистрацией и первой сделкой
- Разрез по источникам привлечения, категорией товаров, объему сделок
- Статистика по использованным промокодам

3.Удержание
- Кол-во клиентов без сделок в течение 6 месяцев, снова совершивших сделку
- Объем коммуникаций с клиентами, кол-во писем/смс/звонков/Push-уведомлений для активации
- Доля клиентов, совершивших сделку от общего кол-ва, на которых была коммуникация

4.Доход
- Доход, руб
- Расход, руб
- Выручка, руб.
- Расходы по категориям
- Разные разрезы, например по источникам привлечения

5.Рекомендация
- Доля новых зарегистрировавшихся клиентов, перешедших по реферальным ссылкам
- Кол-во отправленных ссылок в качестве рекомендаций
- Доля зарегистрированных к отправленным
